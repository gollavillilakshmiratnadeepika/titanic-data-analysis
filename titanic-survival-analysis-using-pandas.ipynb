## Step 1 : Load the Dataset
We load the Titanic dataset  provided by  Kaggle.

```python
# Load the Dataset Titanic
df = pd.read_csv('/kaggle/input/titanic-machine-learning-from-disaster/train.csv')
# Display the first five rows of Dataset
df.head()
```

## Step 2: Understand the data
We check the columns ,data types and missing values of datasets.

```python
# Basic information about the dataset
df.info()
# Statistical summary of dataset
df.describe()
# Check for the missing values
df.isnull().sum()
```

## **Step 3 : Handle the missing values
We fill the missing Age values with the median and the Embarked values with the most common values.

```python
# Fill missing values in Age with median
df['Age'] = df['Age'].fillna(df['Age'].median())
# Fill  missing values in Embarked with most frequent
df['Embarked'] = df['Embarked'].fillna(df['Embarked'].mode()[0])
# Drop Cabin column(too many missing values)
df = df.drop(columns = ['Cabin'])
# Check again for the missing values
df.isnull().sum()
```

## Step 4 : Data Analysis
We analyze survival based on gender,class and age.

```python
#Survival rate  by gender
print(df.groupby('Sex').Survived.mean())
# Survival rate by passenger class
print(df.groupby('Pclass').Survived.mean())
# Average age of survivors vs non survivors
print(df.groupby('Survived').Age.mean())
```

## Step 5 : Key Insights 
From the analysis :
- Females had a higher survival rate than males.
- 1st class passengers survival rate more than 3rd class
- Younger passengers had slightly better survival rates
